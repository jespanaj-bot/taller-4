```python
import gradio as gr

class IllegalArgumentException(Exception):
    pass

class Vendedor:
    def __init__(self, nombre, apellidos, edad):
        self.nombre = nombre
        self.apellidos = apellidos
        self.edad = edad

    def verificarEdad(self):
        if self.edad < 18:
            raise IllegalArgumentException("El vendedor debe ser mayor de 18 años")
        if self.edad < 0 or self.edad > 120:
            raise IllegalArgumentException("La edad no puede ser negativa ni mayor a 120")

    def imprimir(self):
        return (f"Nombre del vendedor: {self.nombre}\n"
                f"Apellidos del vendedor: {self.apellidos}\n"
                f"Edad del vendedor: {self.edad}")

def procesar(nombre, apellidos, edad):
    try:
        edad = int(edad)
        v = Vendedor(nombre, apellidos, edad)
        v.verificarEdad()
        return v.imprimir()
    except ValueError:
        return "Error: la edad debe ser un número entero"
    except IllegalArgumentException as e:
        return "ERROR: " + str(e)

app = gr.Interface(
    fn=procesar,
    inputs=[gr.Textbox(label="Nombre"), gr.Textbox(label="Apellidos"), gr.Textbox(label="Edad")],
    outputs=gr.Textbox(label="Resultado", lines=6),
    title="Clase Vendedor - Validación de Edad (GUI Colab)"
)

app.launch()
```
